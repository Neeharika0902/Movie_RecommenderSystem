# Now want to convert Json to array
but here we have Json in String format. 
So, First
String --> Array
then we will fetch out required field from the array.

--> ast.literal_eval() : used to convert string to array.

```python
# Json to Array conversion for column 'genres'

import ast
def stringTOarray(obj):
    arrayofM = []

    for i in ast.literal_eval(obj):
        arrayofM.append(i['name'])
    return arrayofM
```

```python
newmovie_df['genres'] = newmovie_df['genres'].apply(stringTOarray)
```

```python
newmovie_df.head()
```

```python
# Json to Array conversion for column 'keywords'
import ast
def stringTOarray(obj):
    arrayofM = []

    for i in ast.literal_eval(obj):
        arrayofM.append(i['name'])
    return arrayofM

newmovie_df['keywords'] = newmovie_df['keywords'].apply(stringTOarray)
```

```python
newmovie_df.head()
```

```python
# now from column 'Cast' we want only first three celebrity's name. same as above just we will add If-else condition of it.

def stringTOarray(obj):
    arrayofM = []
    counter = 0

    for i in ast.literal_eval(obj):
        if counter < 3:
            arrayofM.append(i['name'])
            counter +=1
        else:
            break
    return arrayofM

newmovie_df['cast'] = newmovie_df['cast'].apply(stringTOarray)
```

```python
newmovie_df.head()
```

```python
# To fetch only Director's name from the column 'crew', applied IF-ELSE condition

def findingDirector(obj):
    arrayofM = []

    for i in ast.literal_eval(obj):
        if i['job'] == 'Director':
            arrayofM.append(i['name'])
            break
    return arrayofM

newmovie_df['crew'] = newmovie_df['crew'].apply(findingDirector)
```

## Core Logic of Project ##
--> Now we have 'tags' in string format, but we cannot find similarity between texts. So, we will convert test into vectors. From that Vectors, we can find similar(nearest) vector of other movies to recommend movie to the user. This technique known as 'Text Vectorization'.
 
--> Text Vectorization Techniques are : (1) Bags of Words, (2 )TF-IDF, (3) Word to Vec

Here, we will use Bags of Words.

--> Procedure of bags of words:


Combine all the 'tag' column of the row. 
find out Most_Frequent 5000 words.

so now we have 5000 columns for rows. (one word - one column, similarly 5000 words for every movie)

Then,

for all words in the 'tag', we will count the frequency. How many times that particular word is appered in the text(overview/tag) of the movie.
same thing we will do for all rows.
if movie do not contain word from Most_Frequent 5000 word, we will keep it as 0.

# Stop Words Handling
Words which are used just for sentence formation. does not contribute for recommandations. For Example: 'are', 'is', 'in'


```python
# max_feature = how many words do you want to consider as Most_Frequent 

from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features=5000,stop_words='english')


# This will convert each words into vectors. so, we will have vectors for each movies.
vectors_Movies = cv.fit_transform(new_M_df['tags']).toarray()

```

```python
# If you want to print 5000 most_Frequent words
cv.get_feature_names_out()
```

Now, we should consider 'love', 'loved' as a same word.

similarly, 'actor' and 'actors' should be same words.
 for that we will use nltk library.

```python
import nltk
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()
```

# Meaning of Used Function 
tagText.split() : to convert string into array

ps.stem() : used to avoid duplicate words (by using this funcion we can consider 'love', 'loved', 'loving' as a simgle word 'love')

```python
def stem(tagText):
    y = []
    for i in tagText.split():
        y.append(ps.stem(i))

    return " ".join(y)

new_M_df['tags'] = new_M_df['tags'].apply(stem)
```

```python
new_M_df.head()
```

# Again Vectorization after ransforming Data

```python

from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features=5000,stop_words='english')


# This will convert each words into vectors. so, we will have vectors for each movies.
vectors_Movies = cv.fit_transform(new_M_df['tags']).toarray()

```

```python
cv.get_feature_names_out()
```

# Distance Calculation between two vectors
-> If Distance is high then, similiarity is Low || If Distance is low then, similarity is high.

-> Here, we will not calculate ecludiean distant between two vectors as it will be counted distance between tip points of two vectors. Also, for high dimentional data Euclidean is not reliable.

-> We will calculate Cosine distance (theta (angle) between two vectors).

```python
from sklearn.metrics.pairwise import cosine_similarity

# Now we are finding similarity of each movie vector with all movie vectors.
similary_movies = cosine_similarity(vectors_Movies)
```

# Main Code to Recommend a movie

We will find movie_index  -->  will give movie_index as an input to the main function  -->  based on that movie_index will find similar movies  -->  we need to sort the similary score in desc order for high similarity score  -->  then, we will select first 5 movies which are closely similar to the input movie.

Also, Here we have similarity score of movie with each movies. 

Ex. Movie1's similarity score with rest movies. 

So, we need to store the score with the movie_index. 

EX. for movie1 : (2, 0.34) 
which describes movie1 has 0.34 similarity score with movie2.

```python
# This is just for an example to learn each tasks of the Main Code Function

# To find Index of the movie --------------------------------------------------------------
new_M_df[new_M_df['title'] == 'Avatar'].index[0]


# below code will provide sorted similarity score with indexes. ----------------------------
# lambda x:x[1] : describes that we want to sort data based on second column not on first

sorted(list(enumerate(similary_movies[0])),reverse=True, key=lambda x:x[1])

# Now we will select only first 5 similar movies --------------------------------------------
sorted(list(enumerate(similary_movies[0])),reverse=True, key=lambda x:x[1])[1:6]

# Now entire thing we will combine for the main function.
```

```python
# Now here  we will combine all the things in the main function.
def movie_recom(movie_name):
    movie_index = new_M_df[new_M_df['title'] == movie_name].index[0]
    distances = similary_movies[movie_index]
    similar_movieList = sorted(list(enumerate(distances)),reverse=True, key=lambda x:x[1])[1:6]

    for i in similar_movieList:
        # i[0] : bcoz we will get tupple for each rows like (1214, 0.28676966733820225), So we need only value of first column which is Movie_ID   
        # [i[0]].title : to fetch Name of the already found MovieID
        print(new_M_df.iloc[i[0]].title)  
        
```

```python
# calling our main function movie_recom
movie_recom('Avatar')
```
